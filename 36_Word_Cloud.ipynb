```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
data = pd.read_excel("C:\\Users\\ASUS\\Downloads\\Musical_instruments_reviews.xlsx")
```

```python
data.head()
```

```python
#Trying to generate wordcloud using 1st review
text = data.reviewText[0]
wc = WordCloud().generate(text)  # Create and generate a word cloud image
plt.imshow(wc, interpolation='bilinear') #Display the generated word cloud
plt.axis("off")
plt.show()
```

```python
text = '' 
for val in data.reviewText: 
    val = str(val)      #split function has to be considered in str type only
    token = val.split()   #Split every word and save it in list
    for i in range(len(token)):
        token[i] = token[i].lower()
  
      
text += " ".join(token)+" "
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='pink', random_state = 1, collocations = False,
                stopwords = STOPWORDS, 
                min_font_size = 8).generate(text) 
  
                      
plt.figure(figsize = (30, 20), facecolor = None) 
plt.imshow(wordcloud,interpolation = 'bilinear') 
plt.axis("off")  
  
plt.show()
#Here all the informal words are removed as they have split into tokens
```

```python
# Save the image in the img folder:
wordcloud.to_file("wordcloud.png")
```
